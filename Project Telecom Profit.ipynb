## Agregue os dados por usuário

[Agora, como os dados estão limpos, os agregue por usuário e por período para ter apenas um registro dessas informações. Isso vai facilitar muito as próximas análises.]

```python
# Calcule o número de chamadas feitas por cada usuário por mês. Salve o resultado.
df_calls_grouped = df_calls.groupby(['user_id', 'year_month'])['id'].count().reset_index(name='ctn_calls')
df_calls_grouped.head(10)
```

```python
# Calcule a quantidade de minutos gastos por cada usuário por mês. Salve o resultado.
df_sum_calls_grouped = df_calls.groupby(['user_id', 'year_month'])['duration'].sum().reset_index(name='sum_calls')
df_sum_calls_grouped.head(10)
```

```python
# Calcule o número de mensagens enviadas por cada usuário por mês. Salve o resultado.
df_cnt_messages_grouped = df_messages.groupby(['user_id', 'year_month'])['id'].count().reset_index(name='ctn_messages')
df_cnt_messages_grouped.head()
```

```python
# Calcule o volume de tráfego de internet usado por cada usuário por mês. Salve o resultado.
df_sum_internet_grouped = df_internet.groupby(['user_id', 'year_month'])['gb_used'].sum().reset_index(name='sum_gb')
df_sum_internet_grouped.head(10)
```


<div class="alert alert-success">

[Coloque os dados agregados em um DataFrame para que cada registro nele represente o que cada usuário consumiu em um determinado mês.]

```python
# Junte os dados de chamadas, minutos, mensagens e internet com base em user_id e month
df_merged = pd.merge(df_calls_grouped, df_sum_calls_grouped, on=['user_id', 'year_month'], how='outer')
df_merged = pd.merge(df_merged, df_cnt_messages_grouped, on=['user_id', 'year_month'], how='outer')
df_merged = pd.merge(df_merged, df_sum_internet_grouped, on=['user_id', 'year_month'], how='outer')
df_merged.head()
```

```python
# Adicione as informações sobre o plano
df_aux = pd.merge(df_users.drop(columns='year_month'), df_plans, left_on='plan', right_on='plan_name')
df_merged = pd.merge(df_merged, df_aux, on='user_id')
df_merged.head()
```

[Calcule a receita mensal para cada usuário: subtraia o limite gratuito do pacote do número total de chamadas, mensagens de texto e dados; multiplique o resultado pelo valor do plano; adicione o preço mensal dependendo do plano. Importante: isso talvez não seja tão simples e necessite de várias linhas, considerando as condições do plano. Portanto, não se preocupe se você passar algum tempo nesta etapa.]

```python
# Calcule a receita mensal para cada usuário
def calcula_receita(row):
    receita = row['usd_monthly_pay']
    
    if row['sum_calls'] > row['minutes_included']:
        receita = receita + (row['sum_calls'] - row['minutes_included']) * row['usd_per_minute']
        
    if row['ctn_messages'] > row['messages_included']:
        receita = receita + (row['ctn_messages'] - row['minutes_included']) * row['usd_per_minute']
                             
    if row['sum_gb'] > row['gb_per_month_included']:
        receita = receita + (row['sum_gb'] - row['usd_per_gb']) * row['usd_per_gb']
    return receita
                             
```

```python
df_merged['receita'] = df_merged.apply(calcula_receita, axis=1)
```

## Teste hipóteses estatísticas

[Teste a hipótese de que a receita média dos usuários dos planos Ultimate e Surf são diferentes.]

Hipótese nula (H0): 
A receita média dos usuários dos planos Surf e Ultimate é igual.

Hipótese alternativa (H1): 
A receita média dos usuários dos planos Surf e Ultimate é diferente.

[Formule as hipóteses nula e alternativa, escolha o teste estatístico, escolha o valor alfa.]

```python
# Teste as hipóteses
sample_1 = df_merged[df_merged['plan']=='ultimate']['receita']
sample_2 = df_merged[df_merged['plan']=='surf']['receita']

sample_1.var(), sample_2.var()
```

[Teste a hipótese de que a receita média dos usuários da área de NY-NJ difere dos usuários das demais regiões.]

[Formule as hipóteses nula e alternativa, escolha o teste estatístico, escolha o valor alfa.]

Hipótese nula (H0): 
A receita média dos usuários da aréa NY-NJ é igual a dos os usuários das demais regiões.

Hipótese alternativa (H1): 
A receita média dos usuários da aréa NY-NJ é diferente da dos os usuários das demais regiões.

```python
df_merged['is-nynj'] = df_merged['city'].str.contains('NY-NJ')
```

```python
df_receita_media = df_merged.groupby(['user_id', 'is-nynj'])['receita'].mean().reset_index()
sample_1 = df_merged[df_merged['is-nynj']==True]['receita']
sample_2 = df_merged[df_merged['is-nynj']==False]['receita']

sample_1.var(), sample_2.var()
```

```python
# Teste as hipóteses
from scipy import stats as st
import numpy as np

alpha = 0.05 # nivel crítico de significancia estatística.

# Se o valor-p for menor que alpha rejeitamos a hipotese nula.
results = st.ttest_ind(sample_1, sample_2, equal_var=False) # executando um teste.

print('valor-p:', results.pvalue)# extraindo valor-p.

if results.pvalue < alpha: # comparando o valor- com o limiar.
    print('Rejeitamos a hipotese nula')
    
else:
    print('Não podemos rejeitar a hipotese nula')
```

Não há diferença significativa no valor médio das reiceitas testadas no teste de hipóteses logo não podemos rejeitar a hipotese nula.
